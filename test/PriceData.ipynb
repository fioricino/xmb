```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import pandas as pd
import mpld3
mpld3.enable_notebook()
```

```python
# test_switching_sinus

# Симулируем 250 сделок
x = pd.DataFrame(np.arange(250))

# исходная функция - синусоида
source_price_func = lambda x: 12000 + 200*np.sin(0.05*x)

# добавим нормальное распределение
randomized_price_func = np.random.normal(source_price_func(x), 10)

# нарисуем
plt.plot(x, randomized_price_func)
```
